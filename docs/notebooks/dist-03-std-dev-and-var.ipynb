$$Var(X + Y) = Var(X) + Var(Y)$$

THIS PRINCIPLE ONLY HOLDS IF THE X AND Y ARE INDEPENDENT RANDOM VARIABLES<BR>

$$Var(X) = 0.5 * (1 - 0.5) = 0.25\\~\\Var(Y) = 0.167 * (1 - 0.167) = 0.139\\~\\Var(X + Y) = Var(X) + Var(Y) = 0.25 + 0.139 = 0.389$$
