## Transforming Distributions 
Transformation is the process of adding, subtracting some constant value from the measures of central tendency of a distribution. Below you can see the impact (shifting) of adding a constant: <br><br>

### Shifting A Distribution

<br><br>

* lets say k = 1 standard deviation 
* The red distribution is a standard normal distribution
* A shift to the right denotes a positive shift of 1 std dev as depicted with the blue distribution 
* A negative distribution would be a shift to the left 1 std dev
* $\displaystyle \mu_x = 0$
* $\displaystyle \mu_y = 1$
* The range and the IQR stay the same because the distribution is not stretched or shrunk 
 * The range from the start of the distribution to the end of the distribution is the same so range did not change
 * The spread around the median did not change so the IQR is the same
* We see that $\displaystyle \sigma_x = sigma_y$
<br>

### Scaling A Distribution 

Lets say k = 2<br>
We multiply the dataset X by k what happens? <br><br>

<br>
When we multiply X by k, mean, median, mode, range, and IQR are all doubled. This results in the distribution being doubled due to the multiplication of X by k. Keep in mind the area of the distribution cannot change. So the if the length of the distribution doubles, the height of the distribution must be halved. In summary: <br>

* the width of the distribution is 2 X
 * mean, median, mode, range, and IQR are all multiplied by 2
* The standard deviation is scaled by k you as well by multiplying by 2
