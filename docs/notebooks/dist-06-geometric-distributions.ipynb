### Requirements of a Geometric Random Variable

1. Each trial must be independent
2. The outcome of each trial must be either a "success" or "failure" (Bernoulli Distribution)
3. The probability of success must be constant across all trials 

### The Application Of A Geometric Random Variable
A simple example of a Geometric Random Variable is how many times must I shoot a 3-pointer before I have success. We no concrete idea how many shots (trials) it will take, but the point is this is a model of a Geometric Random Variable. You want to capture the probability of the first success. 

### Denoting A Geometric Random Variable 
A Geometric Random Variable is denoted as:<br>

$$\displaystyle P(S = n) = p(1 - p)^{n - 1}$$
<Br><Br>
The equation expresses the first Success (S) to occur on the $n^{nh}$ trial where the probability of that happening is equal to p (The probability of success) times (1 - p) the probability of failure to the power of n - 1 (the number of trial - 1) <br>
<br>
Basically what the formula is saying is multiply the failure rate by itself $(1 -p)^{n - 1}$ until you have a success, then multiply by the success rate p. When you look at this way it becomes clear. Lets say you want to know the probability of success on the 5th trial:<br>

$$P(S = n) = p(1 - p)^{n - 1} \Rightarrow p(S = 5 ) = ~(success)~ \cdot ~(failure)^4$$ 
<br>

### Geometric Less Than Probability 
With the "Less Than" probability, instead of asking for the probability of how many trials are needed before a success, the <i>less than probability</i> is <i>What is the probability of of having a success <b>less than</b> x amount of trials ?</i><br>
<Br>
When you ask the "Less Than" question, such as what is the probability of success in less than 4 trials, what you are really asking is, will I have success on the first, second, or third trial. <b>One of those 3 must be true</b><br>
<br>
So statistically the question is conveyed like this:<br>

$P(S < 4) = P(S = 1) + P(S = 2) + P(S = 3)$ Which is how you would calculate the "less than" probability <br><Br>
Also note that $P(S < 4)$ is the same as $P(S \leq 3)$<Br><br>
So how would you calculate <i>What is the probability of having a success in less than 4 trials given the probability of 0.70 ?</i><br>

```python
import sys
sys.path.insert(0, "..")
import resources.datum as datum 
from IPython.display import display, Math 

data = datum.Data()

statement = 'winning a prize'
success = 0.70
n = 4

geometric_dict = {
    "statement" : statement,
    "n" : n,
    "success" : success 
}
res = []
for i in range(1,4): # first, second, and third trial 
    geometric_dict = {
        "statement" : statement,
        "n" : i,
        "success" : success 
    }
    ans = data.get_geometric_probability(geometric_dict, std_out="Y")
    res.append(ans)
    msg = '\\displaystyle P(S = %s) = %s'
    display(Math(msg%(i,round(ans, 4))))
msg = "\\displaystyle P(S < 4) = %s + %s + %s = %s\\\\~\\\\"
msg = msg + "\\textbf{There is %s pct chance of winning a prize before the 4th trial}"
display(Math(msg%(round(res[0], 4), round(res[1], 4), round(res[2], 4), round((sum(res)), 4), round(100 * sum(res), 4))))
```
