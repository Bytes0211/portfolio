## One And Two Tail Test For Difference Of Mean
<b>Introduction</b><br>
In today's data-driven world, businesses can't rely on intuition alone to make critical decisions. Instead, they need to use statistical techniques like hypothesis testing to back up their actions with solid data. One popular test for this purpose is the t-test, which allows you to compare the means of two groups and determine whether they're statistically different.<br><br>
<b>Objective</b><br>
Use one and two tail test to make inferences about the difference of population<br><br>

### Criteria For One and Two Tail Test For Difference Of Means

#### Understanding Hypothesis Testing For Difference Of Means
As we have learned earlier, the hypothesis statement are based on one of three options.<br>
The Null Hypothesis will be based on:<br>

* Equal To
* Less Than or Equal To
* Greater Than Or Equal To
<br><br>

While the Alternative Hypothesis will be based on the inverse of the Null Hypothesis:<br>

* Not Equal To
* Greater Than
* Less Than. 
<br><Br>

<span style = 'color:skyblue;size:103%'>
The selection of tail test depends on the Hypothesis Statement written
</span><br><br>
In the first line of the table below you can see that if your Null Hypothesis is written with and equal statement, that means your Alternative Hypothesis will have a not equal statement, thus your test will be Two Tail. If your Null Hypothesis is written with less than or equal to, or greater than or equal to, your Alternative Hypothesis will be greater than or less than respectively, with One Tail Upper , and One Tail Lower Test. 
<br><br>


|  $H_0$     |  $H_a$  |      Test      |
|:----------:|:-------:|:--------------:|
| $\mu_1 = \mu_2$ |  $\mu_1 \ne \mu_2$ |    Two Tail    |
| $\mu_1 \leq \mu_2$ | $\mu_1 \gt \mu_2$  | One Tail Upper |
| $\mu_1 \geq \mu_2$ | $\mu_1\lt \mu_2$  | One Tail Lower |

<br><br>
Another way to think about writing hypothesis statements and which tail to apply is directionality. If your hypothesis statement contains "equals" or "not equals" there is no directionality. The outcome is either equal or not. Therefore you should apply a two tail test. If your null hypothesis is "less than" then your alternative hypothesis is greater than, meaning pointing to the right. Then you should apply an upper tail test. If your null hypothesis is "greater than", then your alternative hypothesis is less than, meaning point to the left, then you you should apply a lower tail test

<br><br>

The operators used in the Null Hypothesis are based on your assessment and intuition of the data by observing central tendency parameters such as mean, mode, median, outliers. Its not based on any formula. you decide. And the operators for the Alternative Hypothesis will always be the inverse of the Null Hypothesis 
</span>

<span style = 'color:skyblue;font-weight:bold;size:102%'>

<span style = 'color:gainsboro;font-weight:bold;size:100%'>
The One Tail Upper and One Tail Lower Hypothesis Statements are determined by how you define your populations.
</span><br>
Example:<br>
You hypothesize that men earn more than women. Therefore you define population 1 as men and population 2 as women<br>

$\displaystyle \qquad \mu_1 > \mu_2$<br>
This is an one tail - upper hypothesis statement<br><br>
However by making the men population, population 2, and making the women population, population 1 changes the hypothesis even though the data is the same<br>
$\displaystyle \qquad \mu_1 < \mu_2$<br>
This is an one tail - lower hypothesis statement<br><br>

#### Two Tail Test For Difference Of Means
<b>Hypothesis Statement</b><br>
The basis of a two tail test is that the difference of means will NOT equal zero. Therefore the Null Hypothesis will state the difference of the means will equal zero.<br><br>

* $H_0 \rightarrow \mu_1 = \mu_2$
* $H_a \rightarrow \mu_1 \neq \mu_2$
<br><br>

#### One Tail - Upper Test For Difference Of Means
<b>Hypothesis Statement</b><br>
The basis of a one tail -upper  test is that the mean of sample 1 is greater than the mean of sample 2. Therefore the Null Hypothesis will state that mean of sample 1 is less than or equal to mean of sample 2 <br><br>

* $H_0 \rightarrow \mu_1 \leq \mu_2$
* $H_a \rightarrow \mu_1 \gt \mu_2$
<br><br>

#### One Tail - lower Test For Difference Of Means
<b>Hypothesis Statement</b><br>
The basis of a one tail - lower  test is that the mean of sample 1 is less than the mean of sample 2. Therefore the Null Hypothesis will state that mean of sample 1 is greater than or equal to mean of sample 2 <br><br>

* $H_0 \rightarrow \mu_1 \geq \mu_2$
* $H_a \rightarrow \mu_1 \lt \mu_2$
<br><br><br>

### Formulas When Population Standard Deviations Are Unknown

#### Unequal Variances $(\sigma^2_1 \neq \sigma^2_2)$
<br>

Large Samples $\displaystyle (n_1, n_2 \geq 30)$<br><br>

$\displaystyle \qquad z = \dfrac{(\bar{x}_1 - \bar{x}_2) - (\mu_1 - \mu_2)}{\sqrt{\dfrac{s^2_1}{n_1} + \dfrac{s^2_2}{n_2}}}$<br><br>

Small Samples $\displaystyle (n_1, n_2 \lt 30)$<br><br>

$\displaystyle \qquad t = \dfrac{(\bar{x}_1 - \bar{x}_2) - (\mu_1 - \mu_2)}{\sqrt{\dfrac{s^2_1}{n_1} + \dfrac{s^2_2}{n_2}}}$<br><br>

Degrees of Freedom <br><br>

$df = \dfrac{\bigg(\dfrac{s^2_1}{n_1} + \dfrac{s^2_2}{n_2}\bigg)^2}{\dfrac{1}{n_1 - 1} \cdot \bigg(\dfrac{s^2_1}{n_1}\bigg)^2 + \dfrac{1}{n_2 - 1} \cdot \bigg(\dfrac{s^2_2}{n_2}\bigg)^2 }$<br><br><br>

#### Equal Variances $(\sigma^2_1 = \sigma^2_2)$ and Pooled Variance Usage

<b>Pooled Variance</b> aka combined, or overall variance is a way to estimate common variances when you believe that the different populations have the same or near the same variances. As you can see below, it does add a step to your calculation, but makes the formula easier to read and understand at the very least.<br><br>
If the sample variances are equal or almost equal, then we assume that the population variances are as well. thus we can calculate a pooled variance by combining the two sample variances into one. The formula for <i>pooled variance</i> is:<br><br>

$\displaystyle \qquad S^2_p = \dfrac{(n_1 - 1) \cdot s^2_1 + (n_2 - 1) \cdot s^2_2}{n_1 + n_2 - 2}$
<br><Br>
And therefore <b><i>pooled standard deviation</i></b> is :<Br><br>

$\displaystyle \qquad \color{crimson}S_p \color{defaultcolor} = \sqrt{\dfrac{(n_1 - 1) \cdot s^2_1 + (n_2 - 1) \cdot s^2_2}{n_1 + n_2 - 2}}$
<br><br>

Large Samples $\displaystyle (n_1, n_2 \geq 30)$<br><br>

$\displaystyle \qquad z = \dfrac{(\bar{x}_1 - \bar{x}_2) - (\mu_1 - \mu_2)}{\color{crimson} S_{p} \color{defaultcolor} \cdot \sqrt{\dfrac{1}{n_1} + \dfrac{1}{n_2}}}$<br><br>


Small Samples $\displaystyle (n_1, n_2 \lt 30)$<br><br>

$\displaystyle \qquad t = \dfrac{(\bar{x}_1 - \bar{x}_2) - (\mu_1 - \mu_2)}{\color{crimson} S_{p} \color{defaultcolor} \cdot \sqrt{\dfrac{1}{n_1} + \dfrac{1}{n_2}}}$<br><br>

Degrees of Freedom <br><br>

$\displaystyle \qquad df = n_1 + n_2 - 2$<br><br>
Note that the degrees of freedom for equal and unequal variances are very different. 

Confidence Interval for Difference Of Means With Unknown $\sigma$, small sample size, and $s^2_1 = s^2_2$<br><br>

$\displaystyle \large \qquad \quad (a, b) = (\bar{x}_1 - \bar{x}_2) \pm t_{\dfrac{a}{2}} \cdot s_p \cdot \sqrt{\dfrac{1}{n_1} + \dfrac{1}{n_2}}$<br><br><br>

#### Example
A light bulb company has created a new light bulb that they think will last 30 days longer than the previous model.
When a sample of 20 of the new model of bulbs (sample1) and the 20 of the previous model (sample2) they found that the mean life of the new model was 254 days, and for the previous model the mean life was 205 days. standard deviation of Sample1 is 5 days, and standard deviation of Sample2 was 13 days. The company wants to know what they can conclude at a 5% level of significance. <br><br>

|  New Bulb<br>Sample1   |  Old Bulb<br>Sample2  |    
|:-----------:|:----------:|
| $n_1 = 20$  | $n_1 = 20$ |
| $\bar{x}_1 = 254$| $\bar{x}_1 = 205$  |
| $s_1 = 5$ |  $s_2 = 13$ |

<br>

$H_a: \mu_1 - \mu_2 > 30$<br><br>
$H_0: \mu_1 - \mu_2 \leq 30$<br><br>
$a = 0.05 \rightarrow \text{ confidence level: } 0.95$ <br><br>
Since we are stating $H_a: \mu_1 - \mu_2 > 30$ the test should be classified as one tail - upper t-test<br>






```python

import sys
sys.path.insert(0, '..')
import resources.test as test


def calculate_diff_of_mean_with_unknown_sigma():
    test = 'critical_value'
    type = "diff_of_mean"
    sigma1 = 0
    sigma2 = 0
    tail = 'upper'   
    N = 0                                      
    n_1 = 20
    n_2 = 20
    x_bar_1 = 254
    x_bar_2 = 205
    p_hat = 0
    p = 0
    s1 = 5
    s2 = 13
    var_1 = 0
    var_2 = 0
    hypothesize_diff = 30
    CL = 0.95
    margin_of_error_for_calculating_minimum_n = 2
    
    diff_unknown_sigma = {}
    diff_unknown_sigma['test'] = test 
    diff_unknown_sigma['type'] = 'diff_of_mean'
    diff_unknown_sigma['tail'] = tail
    diff_unknown_sigma['N'] = N
    diff_unknown_sigma['sigma1'] = sigma1
    diff_unknown_sigma['sigma2'] = sigma2
    diff_unknown_sigma['n1'] = n_1
    diff_unknown_sigma['n2'] = n_2
    diff_unknown_sigma['xbar1'] = x_bar_1
    diff_unknown_sigma['xbar2'] = x_bar_2
    diff_unknown_sigma['p_hat'] = p_hat
    diff_unknown_sigma['p'] = p
    diff_unknown_sigma['s1'] = s1
    diff_unknown_sigma['s2'] = s2
    diff_unknown_sigma['hypothesize_diff'] = hypothesize_diff
    diff_unknown_sigma['cl'] = CL
    diff_unknown_sigma['mar_err_min_n'] =   margin_of_error_for_calculating_minimum_n
    return diff_unknown_sigma


diff_test = test.Test()

diff_test.make_critical_test(info = calculate_diff_of_mean_with_unknown_sigma(), statement = False, visual = False)

```

The t test score of the differences of means 6.101 is greater than the hypothesized critical value t score 1.711 therefore we can reject the Null Hypothesis $H_0: \mu_1 - \mu_2 \leq 30$ thus supporting the alternative hypothesis $H_a: \mu_1 - \mu_2 > 30$<br><Br>
Therefore the light bulb company can state with 95% confidence that the new version of the light bulb will last at 30 longer than the previous model. <br><br>
