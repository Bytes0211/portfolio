## Binomial Distributions
The binomial distribution is called binomial, as it has two variables:<br>

1. p - the probability of success
2. q - the probability of failure
<br>
Furthermore, since p is the probability of success and q is the probability of failure, it makes sense that p + 1 = 1
<br>
<br>
If an experiment is done <i>once</i>, then you would have a simple Bernoulli Distribution, however if the experiment was executed <i>several</i> times in a row, then you would have a <b><i>Binomial Distribution</i></b>
<br>Put succinctly a Binomial Distribution is <b>a sequence of identical Bernoulli events</b><br>
<br>
<b>Bernoulli Distribution Notation</b><br>
<br>
&emsp;The upper case B denotes a Binomial Distribution<br>
&emsp;$B~(n, p)$<br>
&emsp;&emsp;B - denotes the Binomial Distribution <br>
&emsp;&emsp;n - denotes the number of trials <br>
&emsp;&emsp;p - denotes the probability of success in each trial<br>
&emsp;Example:<br>
&emsp;&emsp;$X \sim B~(10, 0.6) \Leftarrow \text{Variable X follows a Binomial Distribution with 10 trials and a likelihood of 0.6 to succeed on each individual trial }$<br><br>
Also note we can notate a Bernoulli distribution as a Binomial Distribution with a single trial<br>
<br>

&emsp;$Bern(p) = B~(1, p)$<br>
<br>
<br>

### Binomial Random Variables
In order for a variable x to be a binomial random variable. <br>
* Each trial must be independent 
* Each trial can be called a "success" or "failure" 
* There are a fixed number of trials, 
<br>

The probability of success (referred to as p(success))on each trial is constant<br>
* So if x represents how many heads we get when we flip a coin 10 times: 
* Each trial is independent, because the result of one flip doesn't affect the result of any other flip 
* Each trial can be called a success or failure because "heads" is a success and "tails" is a failure 
* We're flipping the coin 10 times, so there is a fixed number of trials 
* The probability of heads is 50% on each flip of the coin, there for the probability is constant 
<br>

So if x is how many heads we get when we flip a coin 10 times, then we could call x a binomial variable<br>
<Br>
On the other hand lets think about j as how many jacks we get when we pull 3 cards from a deck without replacing the card after each pull<br>

* Each trial is not independent, because result of the first pull affects the result of subsequent pulls. If we get a jack on the first pull, then the probability we get a jack on the second pull is 3/51. But if get something other than a jack on the first pull, hen the probability we get a jack on the second pull is 4/51 
* Each trial can be called a success of a failure because a jack is a success and anything else is considered a failure 
* There are 3 pulls, so there is a fixed number of trials 
* The probability of getting a jack on the first pull is 4/52, but the probability of getting a jack on the second pull changes depending on what we got on the first pull. The probability of pulling a jack in the each trial is not constant. 
<br><br>
Since it fails 2 of the 4 conditions, j cannot be called binomial random variable. <br>
<br>

### Binomial Probability 
In binomial probability questions, we're often asked to figure out the probability that we get an exact number of "success", assuming we perform a specific number of independent trials. The formula for this is:<br>

$$\displaystyle P(\text{k successes in n attempts}) = \dbinom{n}{k}~p^k (1 - p)^(n - k)$$
<br><br>
Where:<br>

* $\text{p is the probability of success}$
* $\text{k is the exact number of times we want success}$
* $\text{n is the total number of independent trials}$
* $\text{The binomial coefficient } \dbinom{n}{k} \text{ is the combination }_nC_k $

 * $\dbinom{n}{k} =~ _nC_k~= \dfrac{n!}{k!(n - k)!} $

<br>

ℹ️ NOTE:<br>
$\text{The notation } \dbinom{n}{k} \text{ is pronounced "n choose k"}$<br><br>

<b>Example</b><br><br>
There are 3 marbles in a bag: 2 are green and 1 is red. We're going to do 5 trials where we pull a marble, note the color, and then replace the marble.<br>

$\displaystyle \color{firebrick} \bigcirc \quad \color{forestgreen} \bigcirc \quad \bigcirc$<br><br>
What is the probability that we get the red marble 3 times out of 5 pulls?<br> 
<br>
And what is the probability that we get green 2 times out 5 pulls?<br><br>

1. Confirm that this is a binomial random variable<br><br>
&emsp;a. Are the trials independent?<br>
&emsp;&emsp;i. Since we are pulling with replacement the trials are independent<br>
&emsp;b. Can the trials be deemed only as success or failures?<br>
&emsp;&emsp;i. Red is success and green is failure so the outcomes are only success or failure <br>
&emsp;c. Are there a fix number of trials? <br>
&emsp;&emsp;i. There are 5 trials, true<br>
&emsp;d. The probability of success on each trial is constant? <br>
&emsp;&emsp;i. Since we are pulling with replacement, the probability is constant, true <br>
Each criteria has been met so this is a binomial <br><br>

$\displaystyle \color{firebrick} \bigcirc \quad \color{forestgreen} \bigcirc \quad \bigcirc$<br><br>

2. We're trying to pull a red marble 3 times in 5 pulls. To solve this problem we need to first figure out how many possible combinations , then we can plug the amount of combinations into out formula <br><br>

$\displaystyle \qquad \dbinom{n}{k} =~ _nC_k~= \dfrac{n!}{k!(n - k)!}$<br><br>
$\displaystyle \qquad \dbinom{5}{3} =~_5C_3 = \dfrac{5!}{3!(5 - 3)!}$<br>



$\displaystyle \qquad P(\text{k successes}) = \dbinom{n}{k}~p^k (1 - p)^(n - k) = \dbinom{n}{k}$<br>






### Bernoulli vs Binomial Distribution
A teacher gives her class a quiz consisting of 10 true/false questions<br>

* Each question is a Bernoulli Event 
 * The Expected Value of the Bernoulli Event is which outcome we expect for a single trial (in this case question)
* The entire Test is a Binomial Event 
 * The Expected Value of the Binomial Distribution is the number of times we expect to get a specific outcome (8 out of 10 questions correct)
<br>

### Mean and Standard Deviation of Bernoulli Distribution and Binomial Distribution 
Lets say you have a gumball machine that has several colors in it. One of the colors is yellow.<br> 
And lets also say we know the proportion of yellow balls in the gumball machine is 60% $P = 0.60$<br><br>

Lets make Y a Bernoulli random variable where the success is a yellow ball picked, and a failure is a not picking a yellow ball<br>

$$
\displaystyle Y = \begin{cases} 
 1, & \text{pick yellow ball} \\ 
 0, & \text{not a yellow ball} 
 \end{cases} 
$$
<br>

We know the mean of Y is equal to P: $\quad \mu_{y} = P = 0.60$<br><br>

We know the standard deviation of Y is $\sigma_{y} = \sqrt{p( 1 - p)} = \sqrt{0.60 * 0.40} = \sqrt{0.24} = 0.489$<br><br>

Now lets create a Binomial random variable that has 10 Bernoulli trials of Y<br>

$$ X = \text{ sum of 10 independent trials 0 Y}$$
<br>

We can calculate the mean of X as $\mu_x = n \cdot p = 10 \cdot 0.60 = 6$<br><br>

We can calculate the std dev of X as $\sigma_x = \sqrt{n \cdot p \cdot (1 - p)}$<br>

$\qquad \sigma_x = \sqrt{10 \cdot 0.60 \cdot 0.40} = \sqrt{2.4001} = 1.5492$

### The Probability Function of Binomial Distribution

If you wish to get the likelihood of getting an given outcome a precise number of times over the course of n trials you need to introduce the Probability Function Of The Binomial Distribution<br>
<br>
Again for Binomial Distribution each trial is a Bernoulli Trial so we express the probability of the desired outcome and the alternative outcome like this:<br>
<br>
&emsp;&emsp;$P(\text{desired outcome}) = p$<br>
<br>
&emsp;&emsp;$P(\text{alternative outcome}) = 1 - p$<br>
<br>
In order to get our favored outcome exactly y many times over the total trials we need to account for the desired outcome <b>and</b> the alternative outcomes n - y many times as well. If we do not account of the alternative outcomes then our probability would be "at least y many times<br>
<Br>

```python

import numpy as np
from bokeh.layouts import column
from bokeh.io import curdoc, show, output_notebook
from bokeh.models import ColumnDataSource, Grid, LinearAxis, Plot, VBar, Div, TeX


output_notebook(hide_banner=True)
curdoc().theme = 'dark_minimal'

N = 6

# 1 - 6 will be x ticks as well
x = np.linspace(1, 6, N)
prob = 0.167 #0.167 = red, 0 = blue
y1 = [0, 0, 0, 0, prob, prob]
y2 = [prob, prob, prob, prob, 0, 0]

source1 = ColumnDataSource(dict(x=x,top=y1,))
source2 = ColumnDataSource(dict(x=x,top=y2,))


p1 = Plot(
    title='Favorable Outcomes\n', width=500, height=400,
    min_border=0, toolbar_location=None)

glyph1 = VBar(x="x", top="top", bottom=0, width=.99, fill_color="dodgerblue")
glyph2 = VBar(x="x", top="top", bottom=0, width=.99, fill_color="firebrick", fill_alpha = 0.6)

p1.add_glyph(source1, glyph1)
p1.add_glyph(source2, glyph2)

# y ticks 
p1.y_range.start = 0
p1.y_range.end = 1


xaxis = LinearAxis()
p1.add_layout(xaxis, 'below')

yaxis = LinearAxis()
p1.add_layout(yaxis, 'left')


# add grids 
p1.add_layout(Grid(dimension=0, ticker=xaxis.ticker))
p1.add_layout(Grid(dimension=1, ticker=yaxis.ticker))


p1.xaxis.axis_label = 'n trials'
p1.yaxis.axis_label = 'Probability'

curdoc().add_root(p1)

div = Div(text=r"""
<p>
$$\color{firebrick} \text{desired outcome y - many times}$$
<p />
<br>
<p>
$$\color{dodgerblue} \text{alternative outcome}$$
</p>
""")

show(column(p1, div))
```


Also keep in mind there <i>may</i> exist more than one way to reach our desired outcome. To account for this we need to determine y out of n many outcomes will be favorable. These are actually the <b>combinations</b> we have already studied<br>
<br>
For example:<br>
If you wished to find the number of ways in which 4 out of 6 trials can be successful, think of it as the same as picking 4 elements out of a sample space $\Omega$ of 6.<br>
<br>
&emsp;$C^6_4$<br>
<br>
For example there are 3 different ways to get heads twice in in 3 coin flips:<br>
<br>
[Tails, Tails, Heads]<br>
[Tails, Heads, Tails]<br>
[Heads, Tails, Tails]<br>
<br>
&nbsp;$C^n_y = C^3_2$<br>
<br>
Therefore the probability function is the product of the number of combinations of picking y many elements out of n times<br>
<br>

$$p(y) = \left(^n_y\right) \cdot p^y \cdot (1 - p)^{n - y}$$
<br>
<br>
Example:<Br>
Say you purchased a single GM stock and based on history a 60% chance your stop will go up and a 40% chance that your stock will go down on any given day. <br> 
<br>
&emsp;$p(📈) = 60%$<br>
&emsp;$p(📉) = 40%$<br>
<br>
<span style = "color:dodgerblue;font-size:101%">
What is the likelihood that the stock price will increase 3 times in a 5 day week? 
</span><br>
<br>
We need to plug in:<br>
&emsp;- 3 for y<br>
&emsp;- 5 for n<br>
&emsp;- 0.6 for p


```python

from IPython.display import display, Math
import math

omega = 5
favorable = 3

comb = math.comb(omega, favorable)
py = 0.6**3
py_disp = round(0.6**3, 3)
mp = (1 - 0.6)**(5 - 3)
mp_disp = round((1 - 0.6)**(5 - 3), 3)
ans = comb * py * mp 
dis_ans = ans * 100

msg = "p(y) = \\left(^n_y\\right) \\cdot p^y \\cdot (1 - p)^{n - y}\\\\~\\\\"
msg = msg + "p(3) = \\left(^5_3\\right) \\cdot 0.6^3 \\cdot (1 - 0.6)^{5 - 3}\\\\~\\\\"
msg = msg + "p(3) = %s \\cdot %s \\cdot %s = %s"
display(Math(msg%(comb, py_disp, mp_disp, ans)))

display(Math("\\color{dodgerblue}\\text{There is a %s pct chance of the GM stock increasing 3 times in a 5 day week }"%{dis_ans}))
```

### Expected Value Of The Binomial Distribution
Other types of distributions have expected values and variances based on the given parameters, just like the Poisson distribution. Recall that the Binomial distribution has parameters n, representing the number of events and p, representing the probability of “success” (or the specific outcome we are looking for occurring).<br>
<br>
Consider the following scenario: we flip a fair coin 10 times and count the number of heads we observe. How many heads would you expect to see? You might naturally think 5, and you would be right! What we are doing is calculating the expected value without even realizing it. We take the 10 coin flips and multiply it by the chance of getting heads, or one-half, getting the answer of 5 heads. And that is exactly the equation for the expected value of the binomial distribution:<br>

$$\text{Expected (number of heads)} = E(X) = n \cdot p$$ 
<br>
<br>
Note that if we were counting the number of heads out of 5 fair coin flips, the expected value would be:<br>

$$5 \cdot 0.5 = 2.5$$
<br>
<br>
It is ok for the expected value to be a fraction or have decimal values, though it would be impossible to observe 2.5 heads<br>
<br>
Let’s look at a different example. Let’s say we forgot to study, and we are going to guess B on all 20 questions of a multiple-choice quiz. If we assume that every letter option (A, B, C, and D) has the same probability of being the right answer for each question, how many questions would we expect to get correct? n would equal 20, because there are 20 questions, and p would equal 0.25, because there is a 1 in 4 chance that B will be the right answer. Using the equation, we can calculate:<br>

$$\text{Expected(number of right answers)} = 20 \cdot 0.25 = 5$$
<br>
<br>
Example: <br>
A certain basketball player has an 85% chance of making a given free throw and takes 20 free throws<br>
<br>



```python
from IPython.display import display, Math

free_throws_taken = 20
exp_per_made = 0.85
ans = free_throws_taken * exp_per_made

msg = '\\text{Player1 is expected to make %s pct. of his free throws}\\\\~\\\\'
msg = msg + "\\text{So if he shoots %s free throws he should make %s}"
display(Math(msg%(exp_per_made, free_throws_taken, ans)))
```

### Variance Of the Binomial Distribution
The variance of the binomial distribution is the spread of the probability distributions with respect to the mean of the distribution. For a binomial distribution having n trails, and having the probability of success as p, and the probability of failure as q, the mean of the binomial distribution is<br>
<br>

* p - probability of success
* q - probability of failure 
* n - number of trials
* $\mu$ - mean 

$$\mu = n \cdot p$$
<br>
The Variance of the binomial distribution is:
<br>

$$\sigma^2 = n \cdot p \cdot q$$


The variance of a single coin flip will be the probability that the success happens times the probability that it does not happen: p·(1-p), or 0.5 x 0.5. Because we have n = 10 number of coin flips, the variance of a single fair coin flip is multiplied by the number of flips. Thus we get the equation:<br>

$$\text{Variance(number of heads)} = n \cdot p \cdot (1 - p)\\~\\\text{Variance(number of heads)} = 10 \cdot 0.5 \cdot (1 - 0.5) = 2.5$$
<br>
Considering the 20 question multiple choice quiz again, The variance around getting an individual question would be:
<br>

$$\text{Variance(number of correct answers)} = 20 \cdot 0.25 \cdot (1 - 0.25) = 3.75$$
<br>

We would expect to get 5 correct answers $(20 \cdot 0.25 = 5)$ but the overall variance of the probability distribution is 3.75
<br>
<br>
Basketball Player A is a 85% free throw shooter. What is the variance of free throws made from 20 shots? 


```python
from IPython.display import display, Math

free_throw_pct = 0.85
free_throw_attempts = 20
var_free_throws = round((free_throw_pct * free_throw_attempts * (1 - free_throw_pct)), 4)

msg = '\\text{Player A free throw variance on %s attempts with a probability of %s is %s}'

display(Math(msg%(free_throw_attempts, free_throw_pct, var_free_throws)))

```

A student has a 98% chance of arriving on time to class. What is the variance of the student arriving late?<br>
<br>
<b>NOTE<b><br>
Since the on time rate is 0.98, that means the late rate is 0.02

```python
from IPython.display import display, Math

late_rate = 0.02
sched_arrive_days = 180
var_late_days = late_rate * sched_arrive_days * (1 - late_rate)

msg = '\\text{StudentA has an on time rate of %s and is scheduled for %s class days}\\\\~\\\\'
msg = msg + '\\text{The variance for days being late is %s}'

display(Math(msg%(late_rate, sched_arrive_days, var_late_days)))
```

#### stats.binom.rvs()
The binom.ppf() function generates an array containing specified number of random values drawn from the given binomial distribution. In the example below, a histogram is plotted to visualize the result.



```python
import numpy as np
from scipy.stats import binom 
from bokeh.layouts import column
from bokeh.models import Div, TeX
from bokeh.plotting import figure, show, output_notebook, curdoc


output_notebook(hide_banner=True)

curdoc().theme ='dark_minimal'

fig = figure(width=670, height=400, toolbar_location=None,
           title="scipy.binom.rvs( ) Distribution")

# creating a seed, never used it this way  - and is works just like a random generator with seed
np.random.seed(1016)


#creating a vector containing 10000
#random values from binomial distribution
sci_y = binom.rvs(20, 0.5, 0, 10000)



# Histogram
bins = np.linspace(np.min(sci_y), np.max(sci_y), 10)
hist, edges = np.histogram(sci_y, density=True, bins=bins)
fig.quad(top=hist
       ,bottom=0 
       ,left=edges[:-1] 
       ,right=edges[1:]
       ,fill_color="dodgerblue" 
       ,alpha = 0.5
       ,line_color="white"
       ,legend_label=f"{len(sci_y)} binom rvs values")

show(fig)
```
